```python
#Question 8
import numpy as np
r = 0.12;
pva = np.zeros(shape=(10,1));
pvb = np.zeros(shape=(10,1));
for n in range(0,6):
    pva[n] = -1000/((1+r/12)**(n+1));
for n in range(0,6):
    pvb[n] = -900/((1+r/12)**(n+1));
npva = np.sum(pva)-1000;
npvb = np.sum(pvb)-1000;
print(npva)
print(npvb)
```
